```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset from the CSV file
df = pd.read_csv("C:/Users/ranje/Downloads/Dataset_CyberCrime_Sean.csv")

# Remove leading/trailing whitespace from column names for consistency
df.columns = df.columns.str.strip()

# Drop rows where the 'City' column is empty, as these are invalid entries
df.dropna(subset=['City'], inplace=True)

# Set the 'City' column as the main index for easier data selection and plotting
df.set_index('City', inplace=True)

# Display the first few rows to confirm it loaded correctly
print("Data loaded and cleaned successfully:")
print(df.head())
```

```python
# --- Region-wise Crime Analysis ---

# Sort the DataFrame by the 'Total' column in descending order
region_wise_crimes = df['Total'].sort_values(ascending=False)

# Select the top 15 regions for a clearer visualization
top_15_regions = region_wise_crimes.head(15)

# --- Create the Plot ---
# Set the style and size for the plot
plt.style.use('seaborn-v0_8-whitegrid')
plt.figure(figsize=(12, 10))

# Create the bar plot using seaborn for a nice aesthetic
sns.barplot(x=top_15_regions.values, y=top_15_regions.index, palette='viridis')

# Add titles and labels for clarity
plt.title('Top 15 Regions by Total Cybercrime Cases', fontsize=16, pad=20)
plt.xlabel('Total Number of Cases', fontsize=12)
plt.ylabel('City / State', fontsize=12)

# Adjust layout and display the plot
plt.tight_layout()
plt.show()

# Print the top 10 regions to the console
print("--- Top 10 Regions by Total Cybercrime Cases ---")
print(top_15_regions.head(10))
```

```python
# --- Crime Type (Motive) Analysis ---

# Sum the values of each crime motive column. We drop 'Total' to avoid double-counting.
crime_types = df.drop(columns=['Total']).sum().sort_values(ascending=False)

# --- Create the Plot ---
# Set the figure size
plt.figure(figsize=(12, 8))

# Create the bar plot using seaborn
sns.barplot(x=crime_types.index, y=crime_types.values, palette='plasma')

# Add titles and labels
plt.title('Total Cybercrime Cases by Motive', fontsize=16, pad=20)
plt.xlabel('Motive / Crime Type', fontsize=12)
plt.ylabel('Total Number of Cases', fontsize=12)

# Rotate the x-axis labels to prevent them from overlapping
plt.xticks(rotation=45, ha='right')

# Adjust layout and display the plot
plt.tight_layout()
plt.show()

# Print the crime type totals to the console
print("\n--- Total Cybercrime Cases by Motive ---")
print(crime_types)
```
